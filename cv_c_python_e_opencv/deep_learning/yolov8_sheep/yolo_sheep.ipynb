# Detecção de Ovelhas com YOLOv8


Link da documentação:
https://docs.ultralytics.com/modes/predict/#sources

```python
import cv2
from ultralytics import YOLO
import os
```

# Carregar modelo YOLOv8

```python
model = YOLO(os.path.join("model", "best.pt"))
```

# Abrir video de teste

```python
cap = cv2.VideoCapture(os.path.join("video", "1.mp4"))
```

```python
while cap.isOpened():
    # Ler o frame do video
    success, frame = cap.read()

    if success:
        # Rodar a inferencia em cada frame com YOLOv8
        results = model.predict(frame, device="0")

        # Vizualizar o frame resultante
        annotated_frame = results[0].plot()
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break o loop se 'q' for pressionado
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
